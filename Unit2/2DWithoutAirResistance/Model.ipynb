```python
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = 0
y = 0
yEuler = y
angle = math.radians(44)  # the angle from the horizontal in radians
velocity = 4.6
deltaTime = 0.01
time = 0
gravitationalAcceleration = -9.81
```

```python
xVelocity = math.cos(angle) * velocity
yVelocity0 = math.sin(angle) * velocity
yVelocityEuler = yVelocity0
```

```python
xList = []
yListEuler = []
yList = []
```

```python
while y >= -1.25:
    time += deltaTime
    xList.append(x)
    yListEuler.append(yEuler)
    yList.append(y)
    yVelocityEuler += gravitationalAcceleration * deltaTime
    x += xVelocity * deltaTime
    yEuler += yVelocityEuler * deltaTime
    y = yVelocity0 * time + 0.5 * gravitationalAcceleration * time ** 2
```

```python
data1 = pd.read_csv("CalibratedData1.csv")
data2 = pd.read_csv("CalibratedData2.csv")
```

```python
plt.plot(xList, yListEuler, color="red")
plt.plot(xList, yList, color="blue")
plt.plot(data1["X Position (m)"], data1["Y Position (m)"], color="green")
plt.title("Height (m) vs. Time (s)")
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.legend(["Euler's Method", "Analytical Model", "Experimental Data 1"])
plt.savefig("Graph.png", dpi=300)
plt.show()
```
