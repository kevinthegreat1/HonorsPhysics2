# Calculating the Spring Constant based on Force Measurements with Static Displacements

```python
# Import Libraries
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
# Import Data
data = pd.read_csv("AverageForceDisplacementData.csv")
```

```python
def linear_eval(displacement, spring_constant):
    return spring_constant * displacement
```

```python
# Linear Fit
spring_constant, _ = curve_fit(linear_eval, data["Displacement (m)"], data["Average Force (N)"])
print(spring_constant[0])
```

```python
plt.plot(data["Displacement (m)"], data["Average Force (N)"], "o")
plt.plot(data["Displacement (m)"], linear_eval(data["Displacement (m)"], spring_constant), label=f"Spring Constant = {spring_constant[0]} N/m")
plt.xlabel("Displacement (m)")
plt.ylabel("Average Force (N)")
plt.legend()
plt.savefig("SpringConstant.png", dpi=300)
plt.show()
```
